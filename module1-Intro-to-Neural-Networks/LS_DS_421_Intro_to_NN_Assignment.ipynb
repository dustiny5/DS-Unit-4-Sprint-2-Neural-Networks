### Input Layer:
* The feature(s) from the dataset that will be initially passed into the neural network model.
### Hidden Layer:
* Cannot be accessed and interacted. The NN does the calculations for us.
### Output Layer:
* Triggered by the activation function to transform the inputs to the desired outcome.
### Neuron:
* Or node. Receives inputs, multiplies each input by its wegiht, sum all of products and passes the ouputs to the next layer
### Weight:
* Initial weights are initially calculated by random then continually adjusted by the loss function with each iteration/epoch 
### Activation Function:
* Decides whether the cell will activate or not. Depending on which activation function is used, then the NN decides how much signal is passed onto the next layer
### Node Map:
* The architecture of the neural network or the layout of the node's path.
### Perceptron:
* A single node of a NN - One input and one ouput

## Write your own perceptron code that can correctly classify a NAND gate. 

| x1 | x2 | y |
|----|----|---|
| 0  | 0  | 1 |
| 1  | 0  | 1 |
| 0  | 1  | 1 |
| 1  | 1  | 0 |

```python
import numpy as np

np.random.seed(42)

# Create a 4 x 2 matrix of inputs - 4 rows and 2 cols
inputs = np.array([
    [0,0,1],
    [1,0,1],
    [0,1,1],
    [1,1,1]
])

# Target outputs - 4 rows and 1 col or 4 x 1 vector
target = np.array([[1], [1], [1], [0]])
```

```python
# Activation function for outputs
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Back propagation - Updating weights
def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1-sx)
```

```python
# Random initial weights
# We are assuming the interaction of features is positive when only using np.random.random() since its range is 0 to 1.
# Starts off from 0 to 1. Double it then from 0 - 2. Subtract by 1 then from -1 to 1.
weights = 2 * np.random.random((3,1)) - 1
weights
```

```python
# Calculate weighted sum of inputs 4x2 dot 2x1
weighted_sum = np.dot(inputs, weights)
weighted_sum
```

```python
# Put into an activation function - sigmoid
# Above .50 is active and below .5 is inactive
activated_output = sigmoid(weighted_sum)
activated_output
```

```python
# Calculate error by taking the difference of the target and output
error = target - activated_output
error
```

```python
# Adjustment - Use loss Function(gradient descent) and back propagation
adjustment = error * sigmoid_derivative(activated_output)
adjustment
```

```python
# Update weights by adjusting the inputs and repeat calcuations- 2x4 dot 4x1 
weights += np.dot(inputs.T, adjustment)
weights
```

```python
# Put it together to iterate more times

for i in range(10000):
    # 1. Calculate weighted sum of inputs and weight
    weighted_sum = np.dot(inputs, weights)
    
    # 2. Sigmoid Activation Function
    activated_output = sigmoid(weighted_sum)
    
    # 3. Calcuate the error - target - a_output
    error = target - activated_output
    
    # 4. Adjustment - Back Propagation
    adjustment = error * sigmoid_derivative(activated_output)
    
    # 5. Update weights by adjusting the inputs
    weights += np.dot(inputs.T, adjustment)

print("Weights after training")
print(weights)

print("Output after training")
print(activated_output)
# Adding a third column to inputs variable and changing the weights from 2 to 3 will get more accurate outpute
## Not sure why this is happening. Maybe 2 features is not enough?
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/ryanleeallred/datasets/master/diabetes.csv')
print(df.shape)
df.head()
```

```python
X = df.iloc[:, 0:10]
X = X.drop(columns='Outcome')
X = X.values
```

```python
# Remeber to reshape or get an error when running the for loop below
y = df.iloc[:,8].values
y = y.reshape(-1,1)
y.shape
```

```python
X.shape
```

```python
X.T.shape
```

```python
# Use the correct weight vector by checking the number of features
weights = 2 * np.random.random((8,1)) - 1
weights.shape
```

```python
for i in range(100000):
    # 1. Calculate weighted sum of inputs and weight # 768x8 dot 8x1 = 768x1
    weighted_sum = np.dot(X, weights)
    
    # 2. Sigmoid Activation Function # 768x1
    activated_output = sigmoid(weighted_sum)
    
    # 3. Calcuate the error - target - a_output # 768x1
    error = y - activated_output

    # 4. Adjustment - Back Propagation # 768x1
    adjustment = error * sigmoid_derivative(activated_output)

    # 5. Update weights by adjusting the inputs # 8x768 dot768x1 = 8x1
    weights += np.dot(X.T, adjustment)
    
print("Weights after training")
print(weights)

print("Output after training")
print(activated_output)
```

```python
# Check accuracy after training
# accuracy = correct / total
correct = 0
for i in range(len(X)):
    if y[i] == activated_output[i]:
        correct += 1
print(correct / len(X))
```
